# Importing H2o

```python
import h2o
import matplotlib.pyplot

from h2o.automl import H2OAutoML

h2o.init()
```

# Reading data

```python
data = h2o.import_file("model_data.csv")
```

# Changing target to categorical

```python
data["stroke"] =(data["stroke"] == 1).ifelse("yes","no")
```

# Splitting data and training h2o

```python
X = data.drop("stroke").columns
y = "stroke"
```

```python
aml = H2OAutoML(max_models=20,seed=7)
aml.train(x=X,y=y,training_frame=data)
```
